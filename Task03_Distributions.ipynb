```python
import QGBx
from QGBx.distributions import *
from QGBx.devices import QiskitAerSimulator



dev = QiskitAerSimulator(shots = 1000)


dist = Exponential(dev, rate = 0.6)



gen = QGBx.Generator(dev, dist)
gen.galton_board(6)
gen.run()
dev.draw_circuit()
QGBx.Visualiser(dist).plot_with_ideal()
```

```python
dist = HadamardQW(dev)



gen = QGBx.Generator(dev, dist)
gen.galton_board(6)
gen.run()
dev.draw_circuit()
QGBx.Visualiser(dist).plot_with_ideal()
```

```python
dist = HadamardQW(dev, type = "Asymmetric_Left")



gen = QGBx.Generator(dev, dist)
gen.galton_board(6)
gen.run()
dev.draw_circuit()
QGBx.Visualiser(dist).plot_with_ideal()
```
